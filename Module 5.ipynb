```python
#Pyplot Tutorial: matplotlib pyplot is a collection of command style functions that makes
# matplotlib work like MATLAB

#Each pyplot function make some change to a figure: e.g.' create a figure, creates a plotting area
#in a figure, plot some lines in a plotting area, decorates the plot with labels, etc.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,6,4]) #yaxis  
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.grid() # grid on
plt.show()
```

```python
#if you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y
# values and automatically generates the x values for you. Since python ranges start with 0, the
# default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3]
```

```python
plt.plot([2,3,6,4], [1,4,9,16]) # xaxis and yaxis  
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.grid() # grid on
plt.show()
```

```python
plt.plot([2,3,6,4], [1,4,9,16], 'ro')#ro mean red o shaped dots
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0, 5, 0.2)
plt.plot(t,'b--', label='t')
plt.plot(t,t**2, 'r--', label='^2')
plt.legend()
plt.show()
```
